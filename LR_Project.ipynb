```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random


seed=42
m=50

#np.random.seed(seed)
X=2*np.random.rand(m,1)

Y=X*6+9+np.random.randn(m,1)*10


fig,ax=plt.subplots(figsize=(8,8))
ax.plot(X,Y,'o')

XM=np.c_[np.ones((m,1)),X]



def gradientD(XM,Y,learning_rate=0.1,iterations=1000):
  theta=np.random.randn(2,1)
  for iteration in range(iterations):
    theta=theta-((2/m)*XM.T.dot((XM.dot(theta))-Y))*learning_rate





  return theta


theta_final=gradientD(XM,Y)

#Y_pred=XM.dot(theta_final)


Y_pred = XM.dot(theta_final)
ax.plot(X, Y_pred, color='red', label='Final Regression Line')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
