```python

import pandas as pd
from matplotlib.pyplot import subplots



```

```python
boston=pd.read_csv('./BostonHousing.csv')

boston[:5]
```

```python
fig,axes=subplots(figsize=(8,8))
x=boston['lstat']
y=boston['medv']
axes.plot(x,y,'o')
```

```python

boston['medv'].describe()
#boston['medv'].describe()


```

```python
Xmean=12.653063
Ymean=22.532806

devx=((boston['lstat']-Xmean)*(boston['medv']-Ymean)).sum()

devx2=((boston['lstat']-Xmean)**2).sum()

b1=devx/devx2
b0=Ymean-b1*Xmean





fig,axes=subplots(figsize=(8,8))
x=boston['lstat']

y=boston['medv']
Y_pred=b0+b1*x
axes.plot(x,y,'o',label='Actual data')

axes.plot(x,Y_pred,label='Regression line')
axes.set_xlabel('LSTAT (% lower status)')
axes.set_ylabel('MEDV (Median home price)')

```
